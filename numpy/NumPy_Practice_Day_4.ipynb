```python
import numpy as np
import pandas as pd
```

```python
# Q1 Create two 3×3 matrices A and B. Compute their dot product using both np.dot() and @. Verify they give the same result.
```

```python
A = np.random.randint(1,10,size=(3,3))
B = np.random.randint(1,10,size=(3,3))
print(A)
print(B)
print(np.dot(A,B))
print(A@B)
print(np.dot(A,B) == A@B)
```

```python
# Q2 Create a 3×3 matrix, compute its determinant and inverse using np.linalg. Verify that A @ inv(A) gives the identity matrix (approximately).
```

```python
C = np.random.randint(1,10,size=(3,3))
print(C)
inv_C = np.linalg.inv(C)
print(inv_C)
np.round(C @ inv_C,10)
```

```python
# Also worth knowing: if the determinant is 0, the matrix has no inverse and np.linalg.inv() will error. 
# Checking it first is good defensive practice.
```

```python
print('Determinant:', np.linalg.det(C))
```

```python
# Q3 Set np.random.seed(42). Generate 1000 random numbers from a standard normal distribution. Print the mean and std — what do you expect them to be close to?
```

```python
np.random.seed(42)
D = np.random.randn(1000)
print('Mean:', np.round(np.mean(D),2))  # Should be close to 0
print('STD:', np.round(np.std(D),2))     # Should be close to 1 
```

```python
# Q4 Simulate rolling a 6-sided die 10,000 times using np.random.choice(). Count how many times each face appeared using np.unique() with return_counts=True.
```

```python
E = np.random.choice(np.arange(1,7,1),size=10000,replace=True)
np.unique(E, return_counts=True)
```

```python
# Q5 Generate a 100×5 array of random data representing 100 samples and 5 features. Compute the mean and standard deviation of each feature (column).
```

```python
F = np.random.randn(100,5)
```

```python
col_mean = F.mean(axis = 0)
col_std = F.std(axis = 0)
```

```python
for i, (mean, std) in enumerate(zip(col_mean, col_std)):
    print(f"Feature {i+1}: mean = {mean:.4f}, std = {std:.4f}")
```
