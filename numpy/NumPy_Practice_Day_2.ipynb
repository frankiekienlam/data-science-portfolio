```python
# Q1 Create a 1D array of 12 elements and reshape it into 3×4, then 2×6, then 4×3. Print each shape.
```

```python
arr_q1 = np.arange(12)
print(arr_q1.reshape(3,4))
print(arr_q1.reshape(2,6))
print(arr_q1.reshape(4,3))
```

```python
# Q2 Create two 3×3 arrays and stack them vertically with vstack and horizontally with hstack. What are the resulting shapes?
```

```python
a = np.zeros((3,3))
b = np.ones((3,3))
c1 = np.vstack([a,b])
print(c1)
print(c1.shape)
c2 = np.hstack([a,b])
print(c2)
print(c2.shape)
```

```python
# Q3 Create a 4×4 array and extract just the top-right 2×2 corner using slicing.
```

```python
arr_q3 = np.arange(16)
arr_q3 = arr_q3.reshape(4,4)
print(arr_q3)
print(arr_q3[:2,2:])
```

```python
# Q4 Create array a = np.arange(9).reshape(3,3). Create a view b = a[1:]. Modify b[0, 0] = 99. What happens to a? Now try with .copy() and explain the difference.
```

```python
a = np.arange(9).reshape(3,3)
b = a[1:]
b[0,0]=99
a
```

```python
a = np.arange(9).reshape(3,3)
b = a[1:].copy() # this is a copy only so any changes on the copy will not impact the original array
b[0,0]=99 
a
```
