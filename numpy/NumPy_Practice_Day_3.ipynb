```python
# Q1 Create a 4×5 array of random integers (use np.random.randint()). 
# Compute the sum, mean, and max along each axis. 
# Explain what axis=0 vs axis=1 gives you.
```

```python
arr_q1 = np.random.randint(0,100,size=(4,5))
print(arr_q1)
```

```python
# Compute the sum, mean, and max along each axis. 
# axis = 0 -> result has shape of one row
print(np.sum(arr_q1,axis=0))
print(np.mean(arr_q1,axis=0))
print(np.max(arr_q1,axis=0))
```

```python
# axis = 1 -> result has shape of one col
print(np.sum(arr_q1,axis=1))
print(np.mean(arr_q1,axis=1))
print(np.max(arr_q1,axis=1))
```

```python
# Q2 Normalize a 1D array of 10 random numbers to the range [0, 1] using only NumPy operations. Formula: (x - x.min()) / (x.max() - x.min()).
```

```python
arr_q2 = np.random.randint(1,100,10)
print(arr_q2)
```

```python
(arr_q2 - arr_q2.min()) / (arr_q2.max() - arr_q2.min())
```

```python
# Q3 Create a 3×3 array and add a 1D array [10, 20, 30] to it using broadcasting. 
# Now try adding a column vector (reshape to (3,1)) and explain the difference.
```

```python
arr_q3 = np.random.randint(1,100,9).reshape(3,3)
print(arr_q3)
```

```python
row = np.array([10, 20, 30])
```

```python
arr_q3 + row # broadcast to each row
```

```python
col = row.reshape(3,1)
print(col)
print(col.shape)
```

```python
arr_q3 + col # broadcast to each col
```

```python
# Q4 Given array scores = np.array([45, 82, 91, 36, 78, 55, 88]), 
# use np.where() to label each score as 'Pass' if >= 60, else 'Fail'.
```

```python
scores = np.array([45, 82, 91, 36, 78, 55, 88])
np.where(scores >= 60, 'Pass', 'Fail')
```

```python
# Q5 Create a 5×5 array of random floats. Subtract the column mean from each column (hint: use axis=0 and broadcasting).
```

```python
arr_q5 = np.random.rand(5,5)
arr_q5
```

```python
# Subtract the column mean from each column (hint: use axis=0 and broadcasting).
```

```python
arr_q5 - np.mean(arr_q5,axis=0)
```

```python
print((arr_q5 - np.mean(arr_q5, axis=0)).mean(axis=0))
# should be all ~0.0 — confirms each column is now mean-centered
```
