```python
import numpy as np
import pandas as pd
```

```python
# Q1 Create a 1D array of numbers 1–20. Slice out only the even-indexed elements.
```

```python
arr_q1 = np.arange(1,21,1)
arr_q1[arr_q1%2==0]
```

```python
# Q2 Create a 3×4 array of zeros, then replace the entire second row with 7s.
```

```python
arr_q2 = np.zeros((3,4))
arr_q2[1,:] = 7
arr_q2
```

```python
# Q3 Create two arrays using np.linspace() — one from 0 to 1 (10 elements) and one from 0 to 100 (10 elements). 
# Print their dtypes and shapes.
```

```python
arr_q3_1 = np.linspace(0,1,10)
print(arr_q3_1)
print(arr_q3_1.shape)
print(arr_q3_1.dtype)
arr_q3_2 = np.linspace(1,100,10)
print(arr_q3_2)
print(arr_q3_2.shape)
print(arr_q3_2.dtype)
```

```python
# Q4 Given the array [3, 1, 4, 1, 5, 9, 2, 6], return all elements greater than 4 using boolean indexing.
```

```python
arr_q4 = np.array([3, 1, 4, 1, 5, 9, 2, 6])
arr_q4[arr_q4>4]
```

```python
# Q5 Create a 4×4 identity-like array where the diagonal is [1, 2, 3, 4] using np.diag().
```

```python
arr_q5 = np.diag([1,2,3,4])
arr_q5
```
