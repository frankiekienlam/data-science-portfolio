```python
import numpy as np
import pandas as pd
```

```python
# Q1 Load titanic.csv into pandas. Extract the Age, Fare, and Survived columns as a NumPy array using .to_numpy(). 
# Check its shape and dtype.
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\Python1\DataSample\titanic.csv')
arr_1 = df[['Age','Fare','Survived']].to_numpy()
print(arr_1)
print(arr_1.shape)
print(arr_1.dtype)
print(arr_1.ndim)
```

```python
# Q2 From that array, separate it into X (Age and Fare columns) and y (Survived column). 
# Print their shapes — this is the exact format scikit-learn expects.
```

```python
X = arr_1[:,0:2]
y = arr_1[:,-1]
print(X.shape)
print(y.shape)
```

```python
# Q3 Using only NumPy on X, compute the mean and std of each column, then manually standardize X to have mean=0 and std=1. This is what StandardScaler does.
```

```python
col_mean = np.nanmean(X, axis=0)
col_std = np.nanstd(X, axis=0)
X_scaled = (X - col_mean) / col_std
print(X_scaled)
print(col_mean)
print(col_std)
```

```python
# Q4 Using np.argmax() and np.argsort(), find the passenger with the highest Fare and list the top 5 passengers sorted by Age.
```

```python
print(X[np.argmax(X[:,1])])
X[np.argsort(X[:,0])[-5:]]
```

```python
# Q5 There are NaN values in Age. Use np.isnan() to find them, 
# then replace them with the column mean — using only NumPy, no pandas.
```

```python
np.isnan(X[:,0])
X[:,0][np.isnan(X[:,0])]=np.nanmean(X[:,0])
```
