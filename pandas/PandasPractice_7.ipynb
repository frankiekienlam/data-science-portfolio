##### Q1 Load time_hours.xlsx and JOIN it to restaurant.csv on the time column. Add the hour range to each restaurant row.
##### Q2 Create a pivot table: rows = day, columns = time (Lunch/Dinner), values = average total_bill.

##### Q4 Which day + time combination generates the highest total revenue (sum of total_bill)? Show a ranked summary table.
##### Q5 Find days where the average tip percentage is below 15% (HAVING equivalent). What pattern do you notice?

```python
import pandas as pd
import numpy as np
df_rest=pd.read_csv(r"C:\Users\user\Desktop\Python1\DataSample\restaurant.csv")
df_time=pd.read_excel(r"C:\Users\user\Desktop\Python1\DataSample\time_hours.xlsx")
print(df_rest.head())
print(df_time.head())
```

```python
# Q1 Load time_hours.xlsx and JOIN it to restaurant.csv on the time column. Add the hour range to each restaurant row.
```

```python
df_q1 = pd.merge(df_rest, df_time, on = 'time')
df_q1.head()
```

```python
# Q2 Create a pivot table: rows = day, columns = time (Lunch/Dinner), values = average total_bill.
```

```python
pd.pivot_table(
    data = df_q1,
    index = 'day',
    columns = 'time',
    values = 'total_bill',
    aggfunc = 'mean' 
)
```

```python
# Q3 Group by smoker and sex. Calculate count of visits, average tip_pct, and average bill_per_person.
```

```python
df_q1['tip_pct'] = df_q1['tip'] / df_q1['total_bill']*100
df_q1['bill_per_person'] = df_q1['total_bill'] / df_q1['size']
```

```python
df_q1.groupby(['smoker','sex']).agg(
    count_of_visits = ('total_bill','count'),
    average_tip_pct = ('tip_pct','mean'),
    average_bill_per_person = ('bill_per_person','mean')
).round(2)
```

```python
# Q4 Which day + time combination generates the highest total revenue (sum of total_bill)? Show a ranked summary table.
df_q4 = df_q1.groupby(['day','time'])['total_bill'].sum().reset_index()
df_q4['rank'] = df_q4['total_bill'].rank(method = 'min', ascending = False).astype(int)
df_q4.sort_values('rank')

# Sat Dinner generates the highest total revenue
```

```python
# Q5 Find days where the average tip percentage is below 15% (HAVING equivalent). What pattern do you notice?
```

```python
avg_tip = df_q1.groupby(['day','time'])['tip_pct'].mean()
avg_tip[avg_tip>15]
# Sat generates the most revenue but the least tip... ppl tend to tip less on bigger bills?
```
