```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\user\Desktop\Python1\DataSample\salaries.csv")
df.shape
df.info()
df.head()
```

```python
# Q1 How many unique job titles are there? What are the top 10 most common job titles?
```

```python
df['JobTitle'].value_counts().reset_index().head(10)
```

```python
# Q2 Calculate average, min, max, and total TotalPay per Year. Which year had the highest average pay?
```

```python
df.groupby('Year').agg(
    average = ('TotalPay', 'mean'),
    min = ('TotalPay', 'min'),
    max = ('TotalPay', 'max'),
    total = ('TotalPay', 'sum')
).reset_index().nlargest(1,'average')
```

```python
# Q3 Find job titles with more than 100 employees AND average TotalPay above $100,000. (HAVING equivalent)
df_q3 = df.groupby('JobTitle').agg(
    emp_cnt = ('Id', 'count'),
    avg_totalpay = ('TotalPay', 'mean')
).reset_index()
df_q3[(df_q3['emp_cnt']>100) & (df_q3['avg_totalpay']>100000)][['JobTitle']]
```

```python
# Q4 Add a dept_avg_pay column showing the average TotalPay for each JobTitle using transform(). Then find employees earning more than 50% above their job title average.
```

```python
df['dept_avg_pay'] = df.groupby('JobTitle')['TotalPay'].transform('mean')
df.head()
df[df['TotalPay'] > df['dept_avg_pay']*1.5][['EmployeeName']]
```

```python
# Q5 Handle nulls in BasePay and Benefits. Fill BasePay nulls with the median, drop rows where TotalPay is null. Report how many rows were affected.
```

```python
BasePay_null = df['BasePay'].isna().sum()
print(f"\nBasePay null count: {BasePay_null}")
df['BasePay'] = df['BasePay'].fillna(df['BasePay'].median())
```

```python
df.dropna(subset=['TotalPay'])
```
