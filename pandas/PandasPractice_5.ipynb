```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\user\Desktop\Python1\DataSample\titanic.csv")
df.shape
df.info()
df.head()
```

```python
# Q1 Load class_name.xlsx. INNER JOIN it to titanic.csv on Pclass = Pclass_number. Verify no rows were lost.
```

```python
class_nm = pd.read_excel(r"C:\Users\user\Desktop\Python1\DataSample\class_name.xlsx")
class_nm.head()
```

```python
df_q1 = pd.merge(df, class_nm, left_on = 'Pclass', right_on = 'Pclass_number')
df_q1.head()
```

```python
# verify no rows were lost after inner join
df.shape[0] == df_q1.shape[0]
```

```python
# Q2 After joining, calculate survival rate (% survived) per ClassName. Which class had the best survival rate?
```

```python
round((df_q1.groupby('ClassName')['Survived'].mean())*100,2) # First class has the highest survival rate
```

```python
# Q3 Split the Titanic data into two DataFrames: survivors (Survived=1) and non-survivors (Survived=0). 
# UNION ALL them back with a new "outcome" label column.
```

```python
# split them up
survivors = df_q1[df_q1['Survived']==True].copy()
non_survivors = df_q1[df_q1['Survived']==False].copy()
```

```python
# union all
survivors['outcome'] = 'survivors'
non_survivors['outcome'] = 'non-survivors'
df_q3 = pd.concat([survivors, non_survivors],ignore_index = True)
```

```python
df_q3.head()
```

```python
# Q4 Using the joined data, find average Fare paid per ClassName and Sex combination.
df_q1.groupby(['ClassName','Sex'])['Fare'].mean().reset_index().round(2)
```

```python
# Q5 LEFT JOIN class_name.xlsx to titanic.csv. 
# Check if any passengers have a Pclass with no matching ClassName. How would you handle this?
```

```python
# left join
df_q5 = pd.merge(df, class_nm, how = 'left', left_on = 'Pclass', right_on = 'Pclass_number')
df_q5.shape[0]
```

```python
# pull records with ClassName equal null
df_q5['ClassName'].isna().sum()
```
