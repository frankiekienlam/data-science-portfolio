```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\user\Desktop\Python1\DataSample\demographic_data.csv")
df.shape
df.info()
df.head()
```

```python
# Q1 Sort countries by Birth rate descending. Show the top 10 highest birth rate countries.
```

```python
df.sort_values(by = 'Birth rate', ascending=0).head(10)
df.nlargest(10,'Birth rate')
```

```python
# Q2 Sort by Internet users descending. Which 5 countries have the highest internet usage?
```

```python
df.sort_values(by='Internet users',ascending = True).tail()
```

```python
df.nlargest(5,'Internet users')[::-1]
```

```python
# Q3 Add a birth_rate_rank column ranking all countries from highest to lowest birth rate.
df['birth_rate_rank'] = df['Birth rate'].rank(method='min',ascending = False)
df.sort_values(by = 'Birth rate', ascending=0).tail(10)
```

```python
# Q4 Find all High income countries with internet usage above 80%. How many are there?
```

```python
df[(df['Internet users']>80) & (df['Income Group'] == 'High income')].shape[0]
```

```python
#Q5 Sort by Income Group ASC, then Internet users DESC within each group. Show the top internet user per income group.
df.sort_values(by='Income Group').groupby('Income Group')['Internet users'].nlargest(1)
```

```python
df.sort_values(by='Income Group').groupby('Income Group').apply(lambda x: x.nlargest(1, 'Internet users'))
```

```python
df.sort_values(by='Income Group').groupby('Income Group').apply(lambda x: x.nlargest(1, 'Internet users'), include_groups = False)
```
