```python
import pandas as pd
import numpy as np
df=pd.read_excel(r"C:\Users\user\Desktop\Python1\DataSample\datetime_demo.xlsx")
print(df.head())
```

```python
# Q1 Load datetime_demo.xlsx. 
# Parse the launched and deadline columns as proper datetimes. 
# Extract year, month, day, and hour from launched into separate columns.
```

```python
df['launched'] = pd.to_datetime(df['launched'])
df['deadline'] = pd.to_datetime(df['deadline'])
df.info()
```

```python
df['launched year'] = df['launched'].dt.year
df['launched month'] = df['launched'].dt.month
df['launched day'] = df['launched'].dt.day
df['launched hour'] = df['launched'].dt.hour
df.head()
```

```python
# Q2 Calculate campaign_duration in days (deadline minus launched). 
# What is the average, min, and max campaign duration?
```

```python
df['campaign_duration'] = (df['deadline'] - df['launched']).dt.days
df.head()
```

```python
df['campaign_duration'].agg(['mean','min','max'])
```

```python
# Q3 The messy_date column contains values like "&10\\201509". 
# Clean it to extract a proper date. (Hint: strip the "&", split on "\\", and reconstruct month + year.)
```

```python
df['messy_date'] = pd.to_datetime(df['messy_date'], format = r'&%m\\%Y%d')
```

```python
df.head()
```

```python
df['new_messy_date'] = df['messy_date'].dt.strftime('%m-%Y')
df.head()
```

```python
# Q4 Group campaigns by launch month (1–12). 
```

```python
df_q4 = df.groupby('launched month').agg({'ID':'size', 'goal':'mean'}).reset_index()
print(df_q4.sort_values('ID', ascending = False))
print(df_q4.sort_values('goal', ascending = False))
```

```python
# Which month sees the most campaign launches? 
print('Month with the most campaign launches:',*df_q4[df_q4['ID'] == df_q4['ID'].max()]['launched month'])
```

```python
# Which month has the highest average goal?
print('Month with the highest average goal:',*df_q4[df_q4['goal'] == df_q4['goal'].max()]['launched month'])
```

```python
# Q5 Find campaigns that launched and had a deadline in the same calendar month. 
# How many are there, and what is their average goal vs campaigns that span across months?
```

```python
# How many are there?
print('# of campaigns that launched and had a deadline in the same calendar month:',df[df['deadline'].dt.month == df['launched month']]['ID'].count())
```

```python
# What is their average goal?
avg_goal_samemonth = df[df['deadline'].dt.month == df['launched month']]['goal'].mean().round(2)
```

```python
# what is average goal campaigns that span across months?
avg_goal_xmonth =df[df['deadline'].dt.month != df['launched month']]['goal'].mean().round(2)
```

```python
# What is their average goal vs campaigns that span across months?
print('Difference b/w their average goal vs campaigns that span across months:', (100-(avg_goal_samemonth / avg_goal_xmonth*100)).round(2), 'smaller')
```
