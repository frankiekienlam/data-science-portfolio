```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\user\Desktop\Python1\DataSample\restaurant.csv")
df.shape
df.info()
df.head()
```

```python
# Q1 Load restaurant.csv. Check for nulls and duplicates. Report your findings.
print('Nulls:\n',df.isna().sum()) # count na 
print('\nDuplicates count:',df.duplicated().sum()) # count duplicates
print('\nDuplicates:\n',df[df.duplicated()]) # pull out duplicates
```

```python
# Q2 Add a tip_pct column: tip as a percentage of total_bill. Round to 2 decimals.
df['tip_pct'] = (df['tip']/df['total_bill']*100).round(2)
df.head()
```

```python
# Q3 Add a tip_category column: "Low" if tip_pct < 10%, "Medium" if 10–20%, "High" if > 20%. Use np.select().
conditions = [
df['tip_pct'] < 10,
df['tip_pct'] <= 20,
df['tip_pct'] > 20 
] 
choices = ['Low', 'Medium', 'High']
df['tip_category'] = np.select(conditions, choices, default = 'Unknown')
df.head()
```

```python
# Q4 Add a bill_per_person column = total_bill / size. Round to 2 decimals.
df['bill_per_person'] = round(df['total_bill'] / df['size'],2)
df.head()
```

```python
# Q5 Filter for weekend dinners (day is Sat or Sun, time is Dinner) with a party size >= 3.
df[(df['day'].isin(['Sat','Sun'])) & (df['time']== 'Dinner') & (df['size']>=3)]
```

```python
# Q5 Filter for weekend dinners (day is Sat or Sun, time is Dinner) with a party size >= 3. USE QUERY
df.query("day in ['Sat','Sun'] and time == 'Dinner' and size >= 3")
```
