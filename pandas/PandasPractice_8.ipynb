```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\user\Desktop\Python1\DataSample\salaries.csv")
display(df.head())
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
# Q1 For each Year, calculate the cumulative total payroll (running total of TotalPay ordered by Year). 
# Show the year-by-year progression.
```

```python
df_q1 = df.groupby('Year')['TotalPay'].sum().reset_index()
df_q1['Running TotalPay'] =  df_q1['TotalPay'].cumsum()
df_q1
```

```python
# Q2 Calculate year-over-year change in average TotalPay using shift(). 
# Which year had the biggest salary jump?
```

```python
# Show the year-by-year progression.
df_q2 = df.groupby('Year')['TotalPay'].mean().reset_index()
df_q2.rename(columns = {'TotalPay':'AvgPay'}, inplace = True)
df_q2['Prev AvgPay'] = df_q2['AvgPay'].shift(1)
df_q2['yoy growth pct'] = (df_q2['AvgPay'] / df_q2['Prev AvgPay'] - 1) * 100
df_q2
```

```python
df_q2['Pct change'] = (df_q2['AvgPay'].pct_change())*100
df_q2
```

```python
print('Year had the biggest salary jump:', df_q1.nlargest(1,'yoy growth pct')['Year'].to_string(index=False))
```

```python
# Q3 Rank employees within each JobTitle by TotalPay (highest = rank 1). 
# Find the top earner in each job title. 
```

```python
df['Rank by JobTitle'] = df.groupby('JobTitle')['TotalPay'].rank(method='min', ascending = False).astype(int)
df.sample(5)
df[df['Rank by JobTitle']==1].sort_values(by='JobTitle')
```

```python
df.groupby('JobTitle').apply(lambda x: x.nlargest(1,'TotalPay'),include_groups=False)
```

```python
# Q4 Add a column showing each employee's pay as a % of their JobTitle's total payroll (percent of total within group).
```

```python
df['Pct of JobTitle Group'] = (df['TotalPay'] / df.groupby('JobTitle')['TotalPay'].transform('sum'))*100
df.sample(10)
```

```python
# Q5 Find employees whose TotalPay increased year over year (hint: use groupby + shift on EmployeeName sorted by Year).
```

```python
df_q5 = df[['EmployeeName','Year','TotalPay']].copy()
df_q5.sample(5)
```

```python
df_q5 = df_q5.sort_values(['EmployeeName', 'Year'])
```

```python
df_q5['Prev TotalPay'] = df_q5.groupby('EmployeeName')['TotalPay'].shift(1)
df_q5.head(10)
```

```python
df_q5[df_q5['TotalPay'] > df_q5['Prev TotalPay']]
```
